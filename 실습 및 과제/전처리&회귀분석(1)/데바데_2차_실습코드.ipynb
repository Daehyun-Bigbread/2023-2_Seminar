# 7-2 이상치 정제하기

```python
df = pd.DataFrame({'sex'   : [1, 2, 1, 3, 2, 1],
                   'score' : [5, 4, 3, 4, 2, 6]})
df
```

```python
# df.value_counts()를 이용해 빈도표 만들어 이상치 확인하기 -> 존재할 수 없는 값 색출
df['sex'].value_counts().sort_index()
```

```python
df['score'].value_counts().sort_index()
```

```python
# sex가 3이면 NaN 부여
df['sex'] = np.where(df['sex'] == 3, np.nan, df['sex'])
df
```

```python
# score가 5보다 크면 NaN 부여
df['score'] = np.where(df['score'] > 5, np.nan, df['score'])
df
```

```python
# 이상치를 결측치로 바꿈 -> 결측치 제거 -> 분석
df.dropna(subset = ['sex', 'score']) \
    .groupby('sex') \
    .agg(mean_score = ('score', 'mean'))
```

```python
# 극단적인 값 제거하기
mpg = pd.read_csv('mpg.csv')

# boxplot으로 극단값 확인하기
import seaborn as sns
sns.boxplot(data = mpg, y = 'hwy')
```

```python
#극단치 기준값
# 1. 1사분위수, 3사분위수
pct25 = mpg['hwy'].quantile(.25)
pct25
```

```python
pct75 = mpg['hwy'].quantile(.75)
pct75
```

```python
# 2.IQR
iqr = pct75 - pct25
iqr
```

```python
# 하한
pct25 - 1.5 * iqr
```

```python
#상한
pct75 + 1.5 * iqr
```

```python
# 3.극단치를 결측 처리하기
# 4.5 - 40.5 벗어나면 NaN 부여
mpg['hwy'] = np.where((mpg['hwy'] < 4.5) | (mpg['hwy'] > 40.5),
                      np.nan, mpg['hwy'])
```

```python
#결측치 빈도 확인
mpg['hwy'].isna().sum()
```

```python
# 4.결측치 제거하고 분석하기
mpg.dropna(subset = ['hwy']) \
    .groupby('drv') \
    .agg(mean_hwy = ('hwy', 'mean'))
```

```python
import pandas as pd
mpg = pd.read_csv('mpg.csv')
```

```python
# x, y축 선정해서 산점도 그리기
import seaborn as sns
sns.scatterplot(data = mpg, x = 'displ', y = 'hwy')
#displ : 배기량
```

```python
# 축 범위 제한하기
sns.scatterplot(data = mpg, x = 'displ', y = 'hwy') \
    .set(xlim = [3, 6], ylim = [10, 30])
```

```python
# hue 사용하여 표식 색 변경
sns.scatterplot(data = mpg, x = 'displ', y = 'hwy', hue = 'drv')
```

# 8-3 막대그래프

```python
# 평균 막대그래프
df_mpg = mpg.groupby('drv', as_index = False) \
    .agg(mean_hwy = ('hwy', 'mean'))

df_mpg
```

```python
sns.barplot(data = df_mpg, x = 'drv', y = 'mean_hwy')
```

```python
# 데이터프레임 정렬 후 막대 그래프 그리기
df_mpg = df_mpg.sort_values('mean_hwy', ascending = False)
sns.barplot(data = df_mpg, x = 'drv', y = 'mean_hwy')
```

```python
# 그룹별로 묶기
df_mpg = mpg.groupby('drv', as_index = False) \
    .agg(n = ('drv', 'count'))

df_mpg
```

```python
sns.barplot(data = df_mpg, x = 'drv', y = 'n')
```

```python
# 빈도 막대 그래프 만들기
sns.countplot(data = mpg, x = 'drv')
```

```python
# 데이터 프레임 변수 값 순서는 f, 4, r -> 고유값
mpg['drv'].unique()
```

```python
# 막대 정렬
sns.countplot(data = mpg, x = 'drv', order = ['4', 'f', 'r'])
```

```python
# drv의 값을 빈도가 높은 순으로 출력
mpg['drv'].value_counts().index
```

```python
# drv 빈도가 높은 순으로 막대 정렬
sns.countplot(data = mpg, x = 'drv',
              order = mpg['drv'].value_counts().index)
```

# 8-4 선 그래프

```python
economics = pd.read_csv('economics.csv')
economics #미국의 여러가지 경제 지표를 월별로 나타낸 데이터
```

```python
sns.lineplot(data = economics, x = 'date', y = 'unemploy')
```

```python
# x축에 연도 표시하기
# 1.날짜 시간 타입 변수 만들기

#날짜가 겹쳐서 굵은 선으로 보이니까 연도가 표시되도록 설정할 것임
#변수 타입을 날짜 시간 타입으로 바꿀 것임
economics['date2'] = pd.to_datetime(economics['date'])
economics.info()
```

```python
# 타입을 바꾸더라도 값이 달라지지는 않는다
economics[['date', 'date2']]
```

```python
# df.dt를 이용한 연도 추출
economics['date2'].dt.year
```

```python
# 월 추출
economics['date2'].dt.month
```

```python
# 일자 추출
economics['date2'].dt.day
```

```python
# 2.연도 변수 만들기
economics['year'] = economics['date2'].dt.year
economics.head()
```

```python
sns.lineplot(data = economics, x = 'year', y = 'unemploy')
```

```python
# 선 아래 표시된 면적(신뢰구간) 제거
sns.lineplot(data = economics, x = 'year', y = 'unemploy', ci = None)
```
